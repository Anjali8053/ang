# Sactter

```python
sns.scatterplot(data= df,x ='age', y = 'fare',hue='sex')
```

# figure size

```python
plt.figure(figsize=(16,10))
sns.scatterplot(data= df,x ='age', y = 'fare',hue='sex')
```

# color change of Palette

```python
plt.figure(figsize=(8,4))
palette = {'male': 'green','female' :'red'}
sns.scatterplot(data= df,x ='age', y = 'fare',hue='sex',palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
```

# save in png and svg 

```python
plt.figure(figsize=(8,4))
palette = {'male': 'green','female' :'red'}
sns.scatterplot(data= df,x ='age', y = 'fare',hue='sex',palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')

plt.savefig('scatter_plot_zz.png' , dpi=300)

plt.savefig('.scatter_plot_zz.svg')
```

# style darkgrid, whitgrid , trick and font style 

```python
sns.set_theme(style='whitegrid',font='arial')

plt.figure(figsize=(8,4))
palette = {'male': 'green','female' :'red'}
sns.scatterplot(data= df,x ='age', y = 'fare',hue='sex',palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
```

# Box Plot

```python
sns.set_theme(style='whitegrid',font='arial',font_scale=1)

plt.figure(figsize=(12,6))

palette = {'First' :'green','Second':'red','Third':'blue'}
sns.boxplot(data=df,x='sex',y='age',hue='class',palette=palette)

plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box plot of Age, based on gender and class of passenger')

plt.legend(loc='upper right' , title='Class Name')
```

# boxen plot

```python
sns.set_theme(style='whitegrid',font='arial',font_scale=1)

plt.figure(figsize=(12,6))

palette = {'First' :'green','Second':'red','Third':'blue'}
sns.boxenplot(data=df,x='sex',y='age',hue='class',palette=palette)

plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box plot of Age, based on gender and class of passenger')

plt.legend(loc='upper right' , title='Class Name')
```

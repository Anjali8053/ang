# line graph

```python
y=np.random.randint(0,10,10)
x=np.arange(0,10)
```

```python
plt.plot(x,y,label='line',color='yellow',linestyle='dashed',linewidth='3',gapcolor='black')
plt.show()
```

```python
dog=('water','drink','chips','amgic','sting','juice')
cat=(400,700,1000,300,5000,600)
```

```python
plt.bar(cat,dog)
plt.show()
```

```python
marks=(700,200,300,500)
students=('ritesh','yash','parveen','samunder')
```

```python
plt.bar(marks,students,color='yellow',edgecolor='blue')
plt.title('students marksheet')
plt.ylabel('student name')
plt.xlabel('marks')
plt.show()
```
